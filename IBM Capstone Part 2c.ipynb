```python
#import Packages

import pandas as pd
import numpy as np

# Folium for creating map cluster
import folium
```

##### Import df from part 2b into part 2c

```python
df = pd.read_csv('./../../Desktop/Toronto3.csv', header=0)

# Drop unwanted column from import
df.drop(columns=['Unnamed: 0'], inplace=True)
```

##### Only use Boroughs that contain Toronto

```python
df1 = df[df['Borough'].str.contains('Toronto',regex=False)]
df1
```

##### Create a map cluster of Toronto neighbrhoods

```python


map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df1['Latitude'],df1['Longitude'],df1['Borough'],df1['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```
